#### 1. 3_sale_price.py 

```python
# This program gets an item's original price and
# calculates its sale price, with a 20% discount.

# Get the item's original price.
original_price = float(input("Enter the item's original price: "))

# Calculate the amount of the discount.
discount = original_price * 0.2

# Calculate the sale price.
sale_price = original_price - discount

# Display the receipt.
print('The original price is', original_price,
      "with 20% discount the sale price is:", sale_price,
      "You have saved", discount)
```

#### example on using try and except

```python
s = input("input: ")
if not isinstance(s, str):
    print(None)
elif not s.isdigit(): # too many digits to convert
    print(None)
else:
    print(int(s))
```

```python
s = input("input: ")
try:
    print(int(s))
except (TypeError, ValueError, OverflowError):
    print(None)
```
